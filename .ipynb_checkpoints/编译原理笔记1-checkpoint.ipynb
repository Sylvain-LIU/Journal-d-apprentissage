# 编译原理笔记1

 1、源程序分析
线性分析（词法分析或扫描）
层次分析 （语法分析）
语义分析

词法分析
作用：把字符流分组为多个记号，每个记号是具有整体含义的字符序列，删除分隔记号的空格。
比如：
position:=initial + rate * 60 被分为：
1.标识符 position
2.赋值符号:=
3.标识符initial
4.加号+
5.标识符rate
6.乘号*
7.数字60

语法分析
作用：获取程序的层次结构（语法树）
比如rate*60是一个有意义的逻辑单元
initial+rate 不能被划分成单个短语

定义语法规则，如表达式的递归定义。

语义分析
任务：检测源程序的语义错误，并收集代码生成阶段要用到的类型信息
主要内容：类型检查——检查每个操作符的操作数是否满足源语言的说明

编译器的各阶段：
源程序——词法分析器——语法分析器——语义分析器——中间代码生成器——代码优化器——代码生成器——目标程序
编译器各个阶段的分组：前端与后端：前端与源语言有关（词法、语法、符号表的建立、语义和中间代码生成和错误处理等）；后端与目标机器阶段有关，如代码优化、生成和错误处理、符号表操作等。

符号表管理器：记录源程序中使用的标识符并收集与每个标识符相关的各种属性信息，包括存储位置、类型、作用域等等
